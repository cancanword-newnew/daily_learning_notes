```python
from coptpy import *
import itertools

env = Envr()

model = env.createModel(name='transportation')

origins,output=multidict({
    'GARY': 1400,
    'CLEV': 2600,
    'PITT': 2900
})

destination,demand=multidict({
    'FRA': 900,
    'DET': 1200,
    'LAN': 600,
    'WIN': 400,
    'STL': 1700,
    'FRE': 1100,
    'LAF': 1000    
})

cost_list = [39,14,11,14,16,82,8,27,9,12,9,26,95,17,24,14,17,13,28,99,20]

orig_desti=tuplelist()

for each in itertools.product(origins,destination):
    orig_desti.append(each)

cost=dict(zip(orig_desti,cost_list))

# 函数详见2025_07_10笔记



num=model.addVars(origins,destination)

for o in origins:
    model.addConstr(quicksum(num[o,d] for d in destination)<=output[o])

for d in destination:
    model.addConstr(quicksum(num[o,d] for o in origins)>=demand[d])
    
model.setObjective(quicksum(num[item] * cost[item] for item in orig_desti),sense=COPT.MINIMIZE)

model.solve()

if model.status == COPT.OPTIMAL:
    # Optimal objective value
    print("Optimal Value of objective:{:.4f}".format(model.objval))
    allvars = model.getVars()
    #Variable value
    print("\nValue of each variable:")
    for var in allvars:
        print("决策变量{0},最优值为 {1:.4f}，基状态为：{2}".format(var.name, var.x, var.basis))
```
