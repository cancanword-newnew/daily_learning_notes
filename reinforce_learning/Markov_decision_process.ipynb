## 一、Markov reward process

```python
import numpy as np
np.random.seed(0)

p=np.zeros((6,6))
p[0,1]=0.1
p[0,0]=0.9
p[1,0]=0.5
p[1,2]=0.5
p[2,3]=0.6
p[2,5]=0.4
p[3,4]=0.3
p[3,5]=0.7
p[4,2]=0.3
p[4,3]=0.5
p[4,1]=0.2
p[5,5]=1.0

def compute_reward(chain,gamma):
    G=0
    for i in reversed(range(len(chain))):
        G = gamma*G+rewards[chain[i]]
        print(G)
    return G
```

```python
rewards=[-1,-2,-2,10,1,0]
gamma=0.5

start_index=np.random.randint(0,len(rewards))
chain=[0,1,2,5]
G=compute_reward(chain,0.5)
```

```python
value=np.dot(np.linalg.inv(np.eye(6,6)-0.5*p),np.array(rewards).T)
```
